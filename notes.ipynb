# SQL Notes


- In postgre SQL if the column names have spaces postgre will give you syntax error if you write them under `SELECT` statement so you need to wrap them in qutation mark like this "Source Tag".

- Difference between `WHERE` and `HAVING`clause, `WHERE` filters individual rows meanwhile `HAVING` filters the groups.

- We have to aggregate columns in `Select` clause if its in `Group By` because those columns have multiple values, if we are aggregating using `AVG()` or other expression or function that result in only one values then it does not need to include in `Group By`

- WHERE asks: "Which row should i keep?"

- CASE asks: "Which value should i put in this row?"

- During applying `COUNT()` don't use `ELSE 0` in the case clause like this:
      ```COUNT(
      CASE
          WHEN condition THEN 1
          ELSE 0
      END
      )
  because `COUNT()` also counts `0` as well
